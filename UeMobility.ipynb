```python
import numpy as np
import matplotlib.pyplot as plt
import os
import re
import pandas as pd

#assuming the files are in the notebook's directory
files = [f for f in os.listdir('.') if re.match(r'day_[0-9]+_UE0_output\.csv', f)]

def date_chars(s):
    start = s.find("day_") + len("day_")
    end   = s.find("_UE")
    substring = s[start:end]
    return(int(substring))

#sort files by day
files = sorted(files, key = date_chars) 

for file in files:
    df = pd.read_csv(file)
    x = df.x 
    y = df.y  
    
    plt.scatter(x, y)
    plt.show()
```
